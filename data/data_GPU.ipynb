```python
import pandas as pd
```

## Création des fichiers CSV des données statiques :

### Table de l’efficacité du matériel

Ce DataFrame rassemble des informations détaillées sur différents types de GPU. Les colonnes sont les suivantes :
- **GPU_Name** : le nom du GPU
  
- **Performance_Tensor** : la performance maximale en TFLOPS pour le traitement tensoriel
- **Max_Power** : la consommation maximale en Watts
- **Release_Date** : la date de sortie du GPU, sous le format "YYYY-MM"
- **CO2_Emissions_Production** : les émissions de CO2 générées lors de la production d'un GPU de ce type, en kilogrammes
- **Water_Consumption_Production** : la consommation d'eau pour la production d'un GPU de ce type en litres
- **Abiotic_Depletion_Production** : l'épuisement abiotiques associé à la production du GPU, exprimé en unités de masse
- **Electronic_Waste_Production** : les déchets électroniques générés lors de la production d'un GPU de ce type, en kilogrammes

```python
df_data_GPU = pd.DataFrame(data = {'GPU_Name': ['V100', 'A100', 'H100'], 
                                   'Performance_Tensor': [125, 500, 2000], 
                                   'Max_Power' : [300, 400, 700], 
                                   'Price' : [26300, 31900, 29500], 
                                   'Release_Date': ['2019-01', '2020-06', '2023-01'],
                                   'CO2_Emissions_Production' : [100, 150, 150], 
                                   'Water_Consumption_Production' : [50, 75, 100],
                                   'Abiotic_Depletion_Production' : [0.015, 0.022, 0.0285],
                                   'Electronic_Waste_Production' : [1.6, 1.6*1.2, 1.6*1.2],
                                   })
df_data_GPU
```

Exportation de la table au format CSV

```python
df_data_GPU.to_csv('csv_files/data_GPU.csv', index=False)
```

### Liste des émissions de CO2 liées à l'utilisation électrique en kgCO2e/kwh par année en France

Ce DataFrame présente les quantités de CO2 émises en kilogrammes équivalent CO2 (kgCO2e) pour chaque kilowattheure (kwh) d'électricité consommé. Les colonnes de ce DataFrame sont les suivantes :

- **Year** : les années
  
- **CO2_per_kwh** : la quantité de CO2 émise par kwh, exprimée en kilogrammes (kgCO2e)

```python
year_2019_2035 = [2019 + i for i in range(17)]
```

```python
df_co2_emissions_per_kwh = pd.DataFrame(data = {'Year' : year_2019_2035, 'CO2_per_kWh' : [0.061, 0.06, 0.057, 0.052, 0.05061538, 0.04923077, 0.04784615, 0.04646154, 0.04507692, 0.04369231, 0.04230769, 0.04092308, 0.03953846, 0.03815385, 0.03676923, 0.03538462, 0.034]})
df_co2_emissions_per_kwh
```

Exportation de la table au format CSV

```python
df_co2_emissions_per_kwh.to_csv('csv_files/co2_emissions_per_kwh.csv', index=False)
```

### Liste de la consommation des [ressources abiotiques par année en kg/Sbeq](https://dataviz.boavizta.org)

Les colonnes de ce tableau sont les suivantes :

- **Year** : les années
  
- **Abiotic_per_kwh** : la quantité ressources abiotiques kilowattheure, exprimée en mètres cubes kg/Sbeq
